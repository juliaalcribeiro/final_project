```python
# Count rows per combination - wine
result_wine = (
    df.groupby(['wine_type', 'wine_category', 'pairing_quality'])
      .size()
      .reset_index(name='count')
      .sort_values(['wine_type', 'wine_category', 'pairing_quality'])
)

print(result_wine)
```

```python
# Combine wine_type + wine_category into one label
result_wine['wine_label'] = result_wine['wine_type'] + ' (' + result_wine['wine_category'] + ')'

# Pivot for plotting
pivot = result_wine.pivot_table(
    index='wine_label',
    columns='pairing_quality',
    values='count',
    fill_value=0
).sort_index()

pivot = pivot.astype(int)

print(pivot)
```

```python
import matplotlib.pyplot as plt

# Plot grouped bar chart
pivot.plot(kind='bar', figsize=(12, 6))

plt.title('Wine Pairing Quality Distribution by Type and Category', fontsize=14)
plt.xlabel('Wine Type (Category)', fontsize=12)
plt.ylabel('Number of Pairings', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Pairing Quality (1–5)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Calculate weighted average rating per wine_label
pivot['avg_rating'] = (
    (pivot[1]*1 + pivot[2]*2 + pivot[3]*3 + pivot[4]*4 + pivot[5]*5)
    / pivot[[1, 2, 3, 4, 5]].sum(axis=1)
)

# Round average rating to 2 decimals for readability
pivot['avg_rating'] = pivot['avg_rating'].round(3)

# Sort wines by highest average rating
pivot_sorted = pivot.sort_values(by='avg_rating', ascending=False)

# Display the top 10 wines with the highest average rating
print(pivot_sorted[['avg_rating']].head(29))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'avg_rating': [3.139, 3.094, 3.092, 3.091, 3.088, 3.083, 3.070, 3.061, 3.048, 3.047, 3.041, 3.041, 3.039, 3.034, 3.032, 3.004, 3.002, 3.002, 2.998, 2.993, 2.991, 2.989, 2.988, 2.984, 2.984, 2.984, 2.888, 2.844, 2.839]
}
index = [
    'Tempranillo (Red)', 'Cabernet Sauvignon (Red)', 'Syrah/Shiraz (Red)', 'Zinfandel (Red)', 'Malbec (Red)',
    'Chardonnay (White)', 'Merlot (Red)', 'Champagne (Sparkling)', 'Sangiovese (Red)', 'Barbera (Red)',
    'Riesling (dry) (White)', 'Cava (Sparkling)', 'Madeira (Fortified)', 'Pinot Noir (Red)',
    'Provence Rosé (Rosé)', 'Ice Wine (Dessert)', 'Sauternes (Dessert)', 'Torrontés (White)', 
    'Chenin Blanc (White)', 'Viognier (White)', 'Albariño (White)', 'Sauvignon Blanc (White)', 
    'Grüner Veltliner (White)', 'Gamay (Beaujolais) (Red)', 'Grenache (Red)', 'Nebbiolo (Red)', 
    'Port (Fortified)', 'Gewürztraminer (White)', 'White Zinfandel (Rosé)'
]
pivot_sorted = pd.DataFrame(data, index=index)


plt.figure(figsize=(10, 10))

# Reduce bar thickness (height=0.4)
plt.barh(
    pivot_sorted.index,
    pivot_sorted['avg_rating'],
    height=0.4,
    color='#555555'   # dark gray
)

plt.gca().invert_yaxis()
plt.title('Wines by Average Pairing Quality', fontsize=12)
plt.xlabel('Average Rating (1–5)', fontsize=12)
plt.ylabel('Wine Type (Category)', fontsize=12)

# **CHANGE 1: Limit the x-axis to "shorten" the bars**
# Set the maximum limit to 3.15 (slightly larger than the maximum value 3.139)
# The minimum limit can be adjusted for focus, here we use 3.0
plt.xlim(2.8, 3.15)


# **CHANGE 2: Bar Labels (inside the plot area)**
for i, v in enumerate(pivot_sorted['avg_rating']):
    # x Position: v - 0.01 (moves the text inside the bar)
    # Color: 'white' (to contrast with the dark bar)
    # Horizontal Alignment: 'right' (anchors the text to the internal edge)
    plt.text(
        v - 0.01,
        i,
        f'{v:.3f}', # Format to 3 decimal places
        va='center',
        ha='right', # Right alignment to keep it inside the bar
        color='white', # White color for contrast
        fontsize=10
    )

plt.tight_layout()
plt.show()
```
